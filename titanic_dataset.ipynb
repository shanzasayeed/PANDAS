```python
data1=pd.DataFrame({'key1':[1,2,4,5,6],
                   'key2':[4,5,6,7,8],
                   'key3':[3,4,5,6,6]
                   })
```

```python
data1
```

```python
data2=pd.DataFrame({'key1':[1,2,45,6,67],
                   'key4':[56,5,6,7,8],
                   'key5':[3,56,5,6,6]
                   })
```

```python
data2
```

```python
pd.merge(data1,data2)   #one column is common. common element is listed along with its other information
```

```python
pd.merge(data1,data2,how='left') #left is data1 ...common column of data1 is taken
```

```python
pd.merge(data1,data2,how='right') 
```

```python
pd.merge(data1,data2,how='outer') #both data is taken
```

```python
pd.merge(data1,data2,how='cross')   #carteductsian pro
```

```python
pd.merge(data1,data2,how='outer')
```

```python
pd.merge(data1,data2,how='outer',on='key1')   #will match on basis of key 1
```

join- same as merge except joins according to rows (index)

```python
data3=pd.DataFrame({'key1':[1,2,4,5,6],
                   'key2':[4,5,6,7,8],
                   'key3':[3,4,5,6,6]},
                   index=['a','b','c','d','e'])
```

```python
data4=pd.DataFrame({'key11':[1,2,4,5,6],
                   'key12':[4,5,6,7,8],
                   'key13':[3,4,5,6,6]},
                   index=['f','g','h','i','j'])
```

```python
data3
```

```python
data4
```

```python
data3.join(data4)  
```

```python
def cat_fare(x):
    if x<10:
        return 'cheap'
    elif x>=10 and x<20:
        return 'mid'
    else:
        return 'high'
```

```python
df['cat_fare']=df['Fare'].apply(cat_fare)
```
